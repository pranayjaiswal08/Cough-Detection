```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
# load the audio file
audio_file_path = 'D:\cough\Cough-4.wav'
signal, sr = librosa.load(audio_file_path)

# calculate the MFCC and Mel spectrogram features
mfccs = librosa.feature.mfcc(signal, sr=sr)
melspec = librosa.feature.melspectrogram(signal, sr=sr)

# plot the MFCC
plt.figure(figsize=(10, 4))
librosa.display.specshow(mfccs, x_axis='time')
plt.colorbar()
plt.title('MFCC')
plt.tight_layout()
plt.savefig('MFCC.png')

# plot the Mel spectrogram
plt.figure(figsize=(10, 4))
librosa.display.specshow(librosa.power_to_db(melspec, ref=np.max), y_axis='mel', fmax=8000, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel spectrogram')
plt.tight_layout()
plt.show()
plt.savefig('Mel.png')
```
