```python
import os
import glob
import numpy as np
import librosa
import soundfile as sf

# Interpolation-based augmentation
def interpolate_augmentation(audio):
    # Drop every second audio sample
    audio_even = audio[::2]
    audio_odd = audio[1::2]

    # Linearly interpolate dropped samples
    audio_even = np.interp(np.arange(len(audio)), np.arange(len(audio_even))*2, audio_even)
    audio_odd = np.interp(np.arange(len(audio)), np.arange(len(audio_odd))*2+1, audio_odd)

    return audio_even, audio_odd

# Add white Gaussian noise augmentation
def noise_augmentation(audio, scale=0.01):
    noise = np.random.normal(scale=scale, size=len(audio))
    audio_augmented = audio + noise
 
    return audio_augmented

```

```python
# Set the path to the directory containing the audio files
audio_dir = 'D:/Noise'

# Loop over all audio files in the directory
for audio_path in glob.glob(os.path.join(audio_dir, '*.wav')):
    print('Processing', audio_path)

    # Load audio data
    audio, sr = librosa.load(audio_path, sr=16000)
    
  # Apply interpolation-based augmentation
    audio_interpolate_even, audio_interpolate_odd = interpolate_augmentation(audio)
 
     # Save augmented audio
    audio_file_base = os.path.splitext(os.path.basename(audio_path))[0]
    sf.write(os.path.join(audio_dir, audio_file_base + '_interpolate_even.wav'), audio_interpolate_even, sr)
    sf.write(os.path.join(audio_dir, audio_file_base + '_interpolate_odd.wav'), audio_interpolate_odd, sr)
    
    # Apply white Gaussian noise augmentation
    audio_noise = noise_augmentation(audio)
    # Save augmented audio
    sf.write(os.path.join(audio_dir, audio_file_base + '_noise.wav'), audio_noise, sr)
```
