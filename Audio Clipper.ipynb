```python
import os
import librosa
import soundfile as sf

# Set the file path and folder path
file_path = 'D:\\aud11.wav'
folder_path = 'D:\\Audio_Data\\seg'

# Load the audio file
y, sr = librosa.load(file_path, sr=16000)

# Set the segment length (in samples)
segment_length = 2*sr  # 1 second

# Iterate over segments and save to folder
for i, start_sample in enumerate(range(0, len(y), segment_length)):
    # Get the end sample for the segment
    end_sample = start_sample + segment_length

    # Ensure the segment doesn't exceed the length of the audio
    if end_sample > len(y):
        break

    # Extract the segment from the audio
    segment = y[start_sample:end_sample]

    # Set the file name for the segment (e.g. file_001.wav)
    file_name = f'file1_{str(i+1).zfill(3)}.wav'

    # Save the segment to the folder
    sf.write(os.path.join(folder_path, file_name), segment, sr)

        # Print the file name
    print(f'Saved {file_name}')
```
